```python
import os
os.environ["SDL_VIDEODRIVER"] = "dummy"
from robotSimulation import RobotSimulation
from IPython.display import HTML
import AOLME
```

```python
if __name__ == '__main__':
    # Make a game instance, and run the game.
       
    """ Type activities for robot movemxents:
        Move forward:  gopi.fw(second)
        Move backward: gopi.bw(second)
        Move left:     gopi.lt(second)
        Move right:    gopi.rt(second)
        Turn left:     gopi.ltd(degree)
        Turn right:    gopi.rtd(degree)
    """
    frame_list = []
    
    red_range   = [50, 255]
    green_range = [0, 90]
    blue_range  = [0, 60]

    detect_color    = (red_range, green_range, blue_range)
    gopi            = RobotSimulation(detect_color, frame_list)
    dis_robot_color = gopi._update_screen()
    collision       = False
    time = True
    
    while(dis_robot_color>10 and time == True):
        
        if collision == False:
            collision, time = gopi.fw(1)
            
        else:
            collision, time = gopi.lt(1)
            collision, time = gopi.bw(1)
            gopi.rtd(90)
            collision, time = gopi.fw(3)
            gopi.ltd(90)
            
            collision = False
        dis_robot_color=gopi._update_screen()
        print(time)
        
    #gopi.run_game()
    
    
    
    all_frame = gopi.frame_list
    print(len(all_frame))
    fps = 30
    play_video = AOLME.vid_show(all_frame, fps) 
    print('Video is ready') 

    HTML(play_video.to_html5_video())
```
