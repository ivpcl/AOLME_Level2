```
from robotSimulation import RobotSimulation
from IPython.display import HTML
import AOLME
from google.colab.patches import cv2_imshow
      
if __name__ == '__main__':
    # Make a game instance, and run the game.
    frame_list = []
    gopi = RobotSimulation(frame_list)
    
    """ Type activities for robot movements:
        Move forward:  gopi.fw(second)
        Move backward: gopi.bw(second)
        Move left:     gopi.lt(second)
        Move right:    gopi.rt(second)
        Turn left:     gopi.ltd(degree)
        Turn right:    gopi.rtd(degree)
    """

   
    gopi.fw(8)
    gopi.ltd(150)
    gopi.lt(5)
    gopi.bw(2)
    gopi.rt(3)
    gopi.ltd(45)
    
    
    all_frame = gopi.frame_list
    #cv2_imshow(all_frame[10])
    print(len(all_frame))
    fps = 5
    play_video = AOLME.vid_show(all_frame, fps)
    


    #gopi.run_game()
    
```
